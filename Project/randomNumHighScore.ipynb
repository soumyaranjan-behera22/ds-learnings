GUESS GAME
store the heighest score in a score file guess game 
 - the heighest score will be the lowest guess ever the user playd

 before playing the game set 1000 in the score.txt 

```python
import random
random_num = random.randint(0,100)
score_count =0
score=0
while True:
    user_num = int(input("enter number"))
    if random_num < user_num:
        print(f"lower than {user_num} ")
    if random_num > user_num:
        print(f"higher than {user_num}")
    if random_num == user_num:
        print("you won ")
        break;
    score_count+=1
    
with open("score.txt", "r")as f:
    heighest_score = f.read()

with open("score.txt","w") as file:
    score = 1000-((score_count-1)*7)
    print(f"old score {heighest_score}")
    if int(heighest_score)<=score:
        heighest_score = score
        file.write(str(score))
print(f"your score is {score} at {score_count} times")
more_score = score - int(heighest_score)
print(f" you won at {more_score} score")
```
