```python
# key - funny
# value = list [....]
print("======== SMART FAKE NEWS HEADLINE GENERATOR ========")
def news_gen(choose):

    templates = [
    "🚨 BREAKING NEWS: {subject} {action} {location}",
    "😱 SHOCKING: {subject} {action} {location}",
    "🔥 VIRAL: {subject} {action} {location}",
    "📢 LATEST UPDATE: {subject} {action} {location}",
    "⚡ JUST IN: {subject} {action} {location}"
    ]

    news_subject = random.choice(categories[choose]["subjects"]).capitalize()
    news_activity = random.choice(categories[choose]["activities"])
    news_location = random.choice(categories[choose]["locations"])
    
    template = random.choice(templates)


    headlines = template.format(
    subject = news_subject,
    action = news_activity,
    location = news_location)
    

    news = f"( {datetime.now()} )\n {headlines}. "
    
    print(news)
    with open ("headline.txt" , "a", encoding = "utf-8") as f:
        f.write(news+"\n")
 
def headline_gen():
    print("===== Headlines On Which Topic ======")
    choose = input("funny/tech/sports").strip().lower()
    if(choose=="funny"):
        news_gen(choose)
    if(choose == "tech"):
        news_gen(choose)
    if(choose == "sports"):
        news_gen(choose)
        
def view_saved_headlines():
    print("===== The Headlines History ===== ")
    with open("headline.txt", encoding = "utf-8")as f:
        show = f.read()
    print(show)


while True:
    choice = input("1. Generate Headlines \n2.View Saves Headlines \n3.Exit")

    match(choice):
        case "1" : headline_gen()
        case "2" : view_saved_headlines()
        case "3" : break
history_save = input("do you want to clear the history").strip().lower()
if(history_save == "yes"):
    with open("headline.txt", "r+") as f:
        f.truncate()
        print("HISTORY DELETED")

print("----------Thank you for using the Headline Generator 👋 --------")
```
