When selecting data in Pandas, sometimes Pandas creates a copy of data and sometimes a view of original data.
This creates confusion called SettingWithCopy conflict.

Simple Explanation
View → changes affect original data
Copy → changes do NOT affect original data

Scalar value means selecting single value from DataFrame.

SettingWithCopyWarning

```python
import pandas as pd

df = pd.DataFrame({
    "marks": [80, 90, 70]
})

new_df = df[df["marks"] > 75]

new_df["marks"] = 100

print(new_df)
print(df)
```

```python

import pandas as pd

df = pd.DataFrame({
    "marks": [80, 90, 70]
})

new_df = df[df["marks"] > 75].copy() # we use copy for the data for mistakely change the values in original data

new_df["marks"] = 100     # here is the copy chages not made in original data set

print(new_df)
print(df)    #---- no changes after the modification in copy
```

#### Grouping and aggregation

```python
# data
data[["country","income"]]
min = data.groupby("country")["income"].min()   # finding min

mean = data.groupby("country")["income"].mean()
print(mean)

# using aggregate function or agg()

aggr = data.groupby("country")["income"].agg(["min","sum","mean","max"])
print(aggr)

agg2 = data.groupby("country").agg({
        "income" : "min",
        "age":"max"
})

agg3 = data.groupby("country").agg(
    avg_val = ("income","mean"),
    avg_age = ("age" , "mean")
    )

agg3
```

#### Melt and pivot

```python
import pandas as pd

# Sample data: Sales revenue per product for Q1 and Q2
data = {
    'Product': ['Laptop', 'Tablet', 'Phone'],
    'Category': ['Electronics', 'Electronics', 'Tech'],
    'Q1_Sales': [12000, 4500, 8000],
    'Q2_Sales': [15000, 5000, 9500]
}

df = pd.DataFrame(data)
# print("Original Wide DataFrame:")
# print(df)

df_long = df.melt(                                     # melting
    id_vars=["Product","Category"],
    value_vars=["Q1_Sales", "Q2_Sales"],
    var_name="Quarter_Sale",
    value_name="Revenue"
)



# pivot
df_short = df_long.pivot(
    index = ["Product","Category"],
    columns = "Quarter_Sale",
    values = "Revenue"
)

print(df_short)
```
