## Series

```python
import numpy as np


# normal declaration
ser = pd.Series([1,2,3,4,5])
print(ser, type(ser))

# numpy arrays declartion
arr2 = np.array([1,2,3,4,5,6])  # numpy arrays creation

ser2 = pd.Series(arr2)  # conversion of numpy  arrays to series
ser3 = pd.Series(arr2 , index=["a","b","c","d","e","f"] )   # adding labels with appropriate indexes here index use d for lebeling the element based on retrival

print(ser2, type(ser2))
print(ser3, type(ser3))
print(ser3["c"], type(ser3))
```

When selecting data in Pandas, sometimes Pandas creates a copy of data and sometimes a view of original data.
This creates confusion called SettingWithCopy conflict.

Simple Explanation
View → changes affect original data
Copy → changes do NOT affect original data

Scalar value means selecting single value from DataFrame.

SettingWithCopyWarning

```python
import pandas as pd

df = pd.DataFrame({
    "marks": [80, 90, 70]
})

new_df = df[df["marks"] > 75]

new_df["marks"] = 100

print(new_df)
print(df)
```

```python

import pandas as pd

df = pd.DataFrame({
    "marks": [80, 90, 70]
})

new_df = df[df["marks"] > 75].copy() # we use copy for the data for mistakely change the values in original data

new_df["marks"] = 100     # here is the copy chages not made in original data set

print(new_df)
print(df)    #---- no changes after the modification in copy
```

```python
import pandas as pd

df = pd.read_csv("data\example.csv")

# apply()
df["tax"] = df["income"].apply(lambda x : "20%" if x>=60000 else "10%" )


# map()
new_gender = {
    "Male" : "M",
    "Female" : "F",
    "Unknown" : "U"
}
df["gender"] = df["gender"].map(new_gender)


# apply()
# new_df = df.assign(column_name1=value1, column_name2=value2, ...)     syntax
# This works because lambda x sees the newly added 'tax' column
df = df.assign(
    tax = df["income"] * 1.1, 
    rate = lambda x: x["tax"] / 10000
)


# replace()
# Replaces 1 with 'A', and 2 with 'B'
# df = df.replace([1, 2], ['A', 'B'])
# # Clean mapping syntax
# df = df.replace({1: 'A', 2: 'B'})
# # Only replaces values inside the 'status' column
# df = df.replace({'status': {'pending': 'hold', 'complete': 'active'}})

df["country"] = df["country"].replace(["USA","Canada","Spain","China","India","Mexico"], ["US", "Ca", "Spn", "Ch", "Ind","Mex"])
df

```
