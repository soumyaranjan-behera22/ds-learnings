🔵 11. Aggregate Functions

Given array:

arr = np.array([[1,2,3],[4,5,6]])
Find:
sum
min
max
Axis-based:
sum of rows
sum of columns
Find index of max value
Difference:
np.sum() vs Python sum()

```python
arr = np.array([[2,4,3],[1,5,6]])
print(np.sum(arr,axis = 1)) # column wise
print(np.sum(arr,axis = 0)) # row wise
print(np.sum(arr)) # overall

min = np.min(arr)
max = np.max(arr)

print(min)
print(max)

max_index = np.where(arr == min)[0][0]  # using conditions

argmin = np.argmin(arr)
argmax = np.argmax(arr)

print(max_index)
print(arr,max_index)
print(argmin)
print(argmax)


```
