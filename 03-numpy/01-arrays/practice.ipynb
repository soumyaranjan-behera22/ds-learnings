```python
# Normalization
arr = np.array([[1,2],[3,4]])
mean = np.mean(arr)
std = np.std(arr)
print(mean)
print(std)

normalization = (arr-mean)/std
print(normalization)

idx_diff = arr-normalization
print(idx_diff)
```
