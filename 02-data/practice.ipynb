```python
import json
def load_data(filename):
    with open(filename, "r") as f:
        data  = json.load(f)
    return data
data = load_data("store_data.json")
def clean_data(data):
    text_to_num = { "one" : 1, "two" : 2, "three" : 3, "four" : 4, None: 10}
    for user in data:

        raw_id = user["id"]

        # agar string hai to strip and lower case
        if isinstance(raw_id, str):
            raw_id = raw_id.strip().lower()

        # convert text to number
        if raw_id in text_to_num:
            raw_id = text_to_num[raw_id]
            user["id"] = raw_id

    return data
clean_data(data)
# print(clean)
```
