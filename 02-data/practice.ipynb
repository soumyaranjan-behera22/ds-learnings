```python
import json
import random
#function to load
def load_data(filename):

    with open(filename, "r") as f:
        data  = json.load(f)
        
    return data
data = load_data("store_data.json")

# function to clean

def remove_inconsistent_data(raw_data):

    # the replacing dictionary with the json data file
    
    text_to_num = { "one" : 1, "two" : 2, "three" : 3, "four" : 4, None: 10}

    # accessing the python obj (key, val) pairs which is "data" by each line like  "id": "one" user is for the single key value pair which calls object interms of json
   #here why we are saying python object the entire json to python converted dictionary "data" is the object 
   #and the every block ex {
    #     "id": "one",
    #     "name": "John Doe",
    #     "age": 28,
    #     "email": "john.doe@gmail.com",
    #     "join_date": "2023-01-15",
    #     "salary": 50000
    # }
    #is called the single object in json

    for user in data:
            # user is for the python key , value
        raw_id = user["id"]

        # agar string hai to strip and lower case
        if isinstance(raw_id, str):
            raw_id = raw_id.strip().lower()

        # convert text to number
        if raw_id in text_to_num:
            raw_id = text_to_num[raw_id]
            user["id"] = raw_id

    return data

def deduplication_as_name(raw_data):
    unique_user = set()  #in this case set is use for storing the unique users and 
    #what should we check in the set 

    #we check that is the user is present in the set or not if not present then append else continue 

    #we slightly modify the checking user as  if the user is present in the set then continue after that add the user["name"] in the set and append the user["name"] in the unique user

    #the condition do that if the item present then skip the loop every time and if not then append and successfully complete the loop
    cleaned_data = [] #this data we have to return 

    for user in data:
        user_name = user["name"].strip().lower()

        if user_name in unique_user:
            continue

        cleaned_data.append(user)
        unique_user.add(user_name)

    return cleaned_data

def deduplication_as_id(raw_data):
    consistent_data = remove_inconsistent_data(data)

    unique_user = set()
    cleaned_data = []

    for user in consistent_data :
        user_id = user["id"]

        if user_id in unique_user:
            continue

        cleaned_data.append(user)
        unique_user.add(user_id)
    return cleaned_data

def missing_values(raw_data):
    for user in data:

        raw_age = user.get("age")

        if(raw_age==None):
            user["age"] = None

    return data

def clean_data(raw_data):
    data =  remove_inconsistent_data(raw_data)
    data = missing_values(raw_data)
    data = deduplication_as_id(raw_data)
    return data
cleaned_data = clean_data(data)   #here is the cleaned data
 
# convert the cleaned data to json
def pre_process_json(cleaned_data):
    with open("pre_process_data.json" , "w") as file:
        json.dump(cleaned_data,file,indent=4)

pre_process_json(cleaned_data)

#ASSIGNMENT 2 -------------
# return average age and
#teenage ratio (if any persons age not given then setthe age as random in between 1 to 18 )

def avg_age(cleaned_data):
    total_age = 0
    total_user = 0
    for user in data:
        total_user+=1
        user_age = user.get("age")
        if(user_age==None):
            user["age"] = 0
        total_age+=int(user["age"])
    return total_age/total_user
        

# deduplication_as_id(data)

# remove_inconsistent_data(data)
# deduplication_as_id(data)

avg_age_user = avg_age(cleaned_data)
if(int(avg_age_user)<avg_age_user<(avg_age_user+1)):
    print(f"avg age users are inbetween {int(avg_age_user)} and {int(avg_age_user)+1} years")

# teenage ratio

def teenage_ratio(cleaned_data):
    for user in data:
        total_user = 0
        teenage_count = 0
        user_age = int(user["age"]);
        if(user_age==0):
            user["age"] = random.randint(1,18)
        if(user_age<18):
            teenage_count+=1;
        print(user)
    return teenage_count/total_user

# teen_ratio = teenage_ratio(cleaned_data)
# print(teen_ratio)


# clean_data(data)
# missing_values(data)
# print(clean)
# print(data)
```
