```python
def large(a,b,c):
    if(a>b and a>c):
        return a
    elif(b>a and b>c):
        return b
    elif(c>a and c>b):
        return c
    else:
        print("all are same")
a=int(input("enter a"))
b=int(input("enter b"))
c=int(input("enter c"))
print(large(a,b,c))
```

```python
#1
def fact(n):
    prod=1
    for i in range(1,n+1):
        prod*=i
    return prod

n= int(input("enter a number"))
fact(n)
# print(fact(n))
```
