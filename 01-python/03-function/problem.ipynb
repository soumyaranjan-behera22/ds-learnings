•If salary < 30,000 → 5%
•If salary is 30,000–70,000 → 15%
•If salary > 70,000 → 25%

Q3. Write a function that prints the digits of a number n, .
For eg: n=3 1 2, there are 3 digits in it 3, 1 and 2 & we need to print them.

AnD COUNT THE DIGITS ALSO

[Hint - The right most digit of a number N is N%10.
And to remove the right most digit from a number, we can do N = N / 10.]

```python
def digit_count(n):
    count=0
    while(n>0):
        dig = n%10
        print(dig)
        count+=1
        n//=10   # strictly integer division
    print(f"the number of digits are {count}")
n=int(input("enter number"))
digit_count(n)
```

Q4. find the sum of digits

```python
def sum_of_dig(n):
    sum=0
    while(True):
        if(n==0):
            break
        else:
            dig=n%10
            n=n//10
            sum+=dig
    print(f"sum of the nums are {sum}")
n=(int(input("enter a number more than 3 digit : ")))
sum_of_dig(n)
```

Q9. Write a function that returns true if n is a prime number and false otherwise, using a loop.

```python
def prime(n):
    count=0;
    for i in range(2,n):
        if n%i==0 :
            count+=1
    if(count==0):
            print( "a prime")
    else:
            print("not prime")
            
prime(70)
        
```
