skip mutiples of 3

```python
i=1
while(i<=10):
    # i+=1
    if(i%3==0):
        i+=1
        continue
    print(i)
    i+=1
```

trafic light system by match case

```python
button = input("enter the color : ")
match button:
    case "red" :
        print("stop")
    case "yellow" :
        print("look")
    case "green" :
        print("go")
    case _:
        print("wrong value")
```
