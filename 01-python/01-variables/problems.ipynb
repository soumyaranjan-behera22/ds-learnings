Q10. Take a decimal number as input (like 45.78) and output its:
•integer part - 45
•fractional part - 45.78

```python
num =float(input("Enter a decimal number"))
int_part = int(num)
# int_num=float(int_part)
raw_float_part = num - int_part
print("integer part is ",int_part)
float_part = int(raw_float_part * 100 + 0.5) / 100.0 #here is the formulla for finding the float part
print("float part is ",float_part)
```
