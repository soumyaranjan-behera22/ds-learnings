Q1. Student Class
private marks
public name
set_marks()
get_marks()
grade()

```python
class Student:
    def __init__(self ,name, marks):
        self.name = name
        self.__marks = marks   # it cant be change due to the user made changes
    def set_marks(self,marks):
        self.__marks = marks
    def get_marks(self):
        return self.__marks
    def grade(self):
        if self.__marks> 90:
            return "A+"
        if 80<= self.__marks<90 :
            return self.__marks," A"
        elif 70<= self.__marks<80 :
            return self.__marks," B"
        elif 60<= self.__marks<70 :
            return self.__marks," C"
        elif 70<= self.__marks<60 :
            return self.__marks," D"
        elif self.__marks < 60:
            return "fail"
        else:return " invalid marks "
s = Student("soumya", 64)
s.grade()
s.set_marks(92)
s.get_marks()
```
