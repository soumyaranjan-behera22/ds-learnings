```python
class Student:
    def __init__(self ,name, marks):
        self.name = name
        self.__marks = marks   # it cant be change due to the user made changes
    def set_marks(self,marks):
        self.__marks = marks
    def get_marks(self):
        return self.__marks
    def grade(self):
        if self.__marks> 90:
            return "A+"
        if 80<= self.__marks<90 :
            return self.__marks," A"
        elif 70<= self.__marks<80 :
            return self.__marks," B"
        elif 60<= self.__marks<70 :
            return self.__marks," C"
        elif 70<= self.__marks<60 :
            return self.__marks," D"
        elif self.__marks < 60:
            return "fail"
        else:return " invalid marks "
s = Student("soumya", 64)
s.grade()
s.set_marks(92)
s.get_marks()
```

```python
class Vector:
    def __init__(self,x,y):
        self.x = x
        self.y = y
    def __add__(self,other):
        return Vector(self.x + other.x, self.y + other.y)

v1 = Vector(2,3)
v2 = Vector(4,1)
v3 = v1+v2
print(v3.x, v3.y)
```

```python
# 33️⃣ Create abstract class Employee
# Method:
# calculate_salary()
# Implement in:
# FullTimeEmployee
# PartTimeEmployee
from abc import ABC, abstractmethod
class Employee(ABC):
    def __init__(self, salary):
        self.salary = salary
    def calculate_salary(self):
        pass
class PartTimeEmployee(Employee):
    def calculate_salary(self):
        # super().__init__()
        return self.salary*1.2
class FullTimeSalary(Employee):
    def calculate_salary(self):
        # super().__init__(salary)
        return self.salary*0.5
e = PartTimeEmployee(20_000)

e.calculate_salary()
e.calculate_salary()

```

```python
# 34. Database System
# Create abstract class Database
# Methods:
# connect()
# disconnect()
# Implement in:
# MySQL
# MongoDB

from abc import ABC ,abstractmethod
class Database(ABC):
    @abstractmethod
    def connect(self):
        pass
class MySQL(Database):
    def connect(self):
        print("connect to MySQL")
    def disconnect(self):
        print("MySQL disconnected ")
class MongoDB(Database):
    def connect(self):
        print("connect to MongoDB")
    def disconnect(self):
        print("MongoDB disconnected ")
sql = MySQL()
sql.connect()
sql.disconnect()
mongo = MongoDB()
mongo.connect()
mongo.disconnect()
```

```python
# 38. Singleton Pattern
# Create class Logger
# Implement Singleton Design Pattern.
class Logger:
    _login = None
    def __new__(cls):
        if cls._login == None:
            print("new user Found")
            cls._login = super(Logger , cls).__new__(cls)
        return cls._login
    def log(self,message):
        print(f"Log: {message}")
logg =Logger()
logg2 =Logger()

logg.log("Program started")
logg2.log("Program END")
```
