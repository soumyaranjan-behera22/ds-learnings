Mthods in oops 
1---> class methods ( cls parameter with the decorator @classmethod)          used for work with class level data 
2----> istance method  (simple method with self and more that one parameter)   acess both class and instance attributes

```python
class Student :
    college = "nalanda"
    def change_clg(self, name):
        print("instance meth and before change ", Student.college)
        self.college = name
        print("instance meth and before change ", Student.college)       # here there is no chaanges made after changing the college name in the instance method
        print(self.college) # for testing
    @classmethod
    def change_scool(cls, name):
        print("before changeing ", Student.college)
        cls.college = name
        print("after changing ",Student.college)  #it changes the name after passing the name through the class method which made changes on the class attribute in the case of instance method that cant be changes 
st = Student()
st.change_clg("ASBM")
st.change_scool("gift")
print("after changing through the object ",st.college)


#simple funda is if we want to chage in the class attribute so we use the @classmethod decorator with cls refrence ---> it cant ,aade changes inthe instance class
#if we work with instace attribute so we use instance attributes 
```
