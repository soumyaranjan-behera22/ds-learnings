```python
# Create a program that:
# Q3.
# 1.Has a list of numbers:[5, 10, 15, 20, 25]
# 2.Uses a list comprehension to create a new list with only numbers greater than 15
# 3.Prints the new list
num = [5,10,15,20,25]
greater15 = [i for i in num if i>15]
print(greater15)
```
