Dictionary  ---- word ---->mean pairs
1.Dictionary Keys must be unique
2.Dictionary Keys must be immutable (e.g., strings, numbers, tuples)
3.Dictionary is mutable.
4.Dictionary values can be anything (lists, other dictionaries, etc.)
5.Unordered (although in modern Python, dictionaries preserve insertion order)

methods are .key(), .values(), .item(), .get(key), .update(key, value)

```python
student = {
    "name" : "soumya",
    "age" : 20,
    "cgpa" : 9.02,
    "pass" : True,
    3.14 : "Btech"
}
for i in student:
    print(i) # it only prints the key

for key,value in student.items():  #the student.items that acces all the elements so that we define and print in key and value way
    print(key,value)

print(student) #this will print as it is the dictionary
```

A set is an unordered collection of unique elements. Example:
    Characteristics
1.Sets can only have unique elements.
2.They are unordered - no indexing or slicing.
3.They are mutable - we can add or remove elements.
4.Set elements must be immutable (like strings, numbers, tuples).

Sets are often used when we need uniqueness or mathematical set
operations.

Set Methods
1. add(val) - adds an element to set
2. remove(val) - removes an element (raises error if not found)
3. clear() - removes all elements
4. pop() - removes and returns a random element (since sets are unordered)
5. s1.union(s2) - returns new union (union is collection of all unique values in both sets)
6. s1.ntersection(s2) - returns new union (intersection is collection of all common & unique values in both sets)

```python
my_set= {1,2,3,1,2,7}
my_set1 ={1,2,3,4,5,6,8}
print(my_set)

print(len(my_set))
print(my_set.add(12))
print(my_set)
my_set.remove(2)
print(my_set)
# my_set.clear()  # it return empty set which is set()
print(my_set) 

my_set.pop()
print(my_set)
my_set.pop()
print(my_set)

print(my_set.union(my_set1))
print(my_set)

print(my_set.intersection(my_set1))

```
