Givean a list of tuples with info(name, subject)
    list all unique courses
    list the students enrolled in english
    create a dictionary (student, set of course)

```python
info = [    
    ("alice" , "math"),
    ("bob", "sci"),
    ("alice","sci"),
    ("charlie","math"),
    ("bob","math"),
    ("alice", "eng"),
    ("charlie","eng"),
]
# print(info)
```

```python
#all the unique course
empty_set= set()
for i in info:
    empty_set.add(i[1])
print(empty_set)
```

```python
#list the students enroll in english

for i in info:
    if(i[1]=="eng"):
        print(i[0])

# or
print("using another method")
for name, course in info:
    if(course=="eng"):
        print(name)
```

```python
# create a dictionary (student, set of course)
dict={}

for name, course in info:
    if(dict.get(name) == None):
        dict.update({name:set()})
        dict[name].add(course)
    else:
        dict[name].add(course)
print(dict)
        
```
