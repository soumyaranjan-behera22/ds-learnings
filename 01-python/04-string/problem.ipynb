Q4.Given a tuple of integers, create:

•A tuple of all even numbers
•A tuple of all odd numbers

```python
tup =(1,2,23,12,7,8,4,3)
for i in tup:
    if(i%2==0):
        print(f"even is {i}")
    else:
        print(f"odd is {i}")
```

```python
even = [x for x in tup if x%2==0]
odd =[x for x in tup if x%2!=0]
print(f"evens are {even}")
print(f"odds  are {odd}")          #this method iss called list comprehension
```

Write a menu-based program where user presses a key (’ A , ‘B , ‘C , ‘D ) depending on the operation they want to perform on the dictionary:
1.A - Add a student
2.B - Update marks
3.C - Search for a student
4.D - Display all students and marks

```python
dic = {}
def add_stu():
    name= input("enter name : ").strip()
    # if(dic[name]==None):   i dont use this bcz it gives error in case of empty or no student records
    if dic.get(name)== None:
        marks= input("enter marks : ")
        dic.update({name:marks})
    else:
        print("Student already exist")
def update():
    name =input ("enter name : ").strip()
    # for key, val in dic.items():
    update_marks=""

    if(dic.get(name)!=None):
        update_marks=input("enter marks : ").strip()
        dic.update({name:update_marks})
    else:
        print("No student records !!")
def search():
    name=input("name for search").strip()
    # if dic[name]!=None : dont use that because it gives error if the elemeent is not exist
    if dic.get(name)!=None:
        print(f"{name} is found...")
    else:
        print(f"{name} Not Exist..")
def display():
    print("Records of the students")
    if dic.items()==None:
        print("No records !!")
    else:
        for key,value in dic.items():
             print(f"{key} : {value} ")
while (True):
    choice = input("Choose option (A = Add, B = Update, C = Search, D = Display , Q = quit ): ").strip().upper()
    #for knowing the methods check the notes
    match choice:
        case "A" :
            add_stu()
        case "B":
            update()
        case "C" :
            search()
        case "D":
            display()
            # break
        case "Q":
            break
```

Q8. Write a program to check whether two lists share no common elements.
    #share no common elemenhts list1 = [1,2,3,4],  list2 =[5,6,7,8]
    #share common elements lis1 = [1,2,3] , list2 =[3,4]
    [hint - use set]

```python
l1 = set([1,2,3,4])
l2 = set([9,5,6,7])
res = l1.intersection(l2)
print(res)
if(res==set()):
    print("no common")
else:
    print("common elements")
```

Q9.    Given a list, print all elements that appear more than once in the list.
[hint- use sets]

```python
li = [1,2,3,1,4,5,2]
li1 = []
for i in li:
    if(li.count(i)>1):
        li1.append(f"{i} :{li.count(i)}")
print(list(set(li1)))
```
