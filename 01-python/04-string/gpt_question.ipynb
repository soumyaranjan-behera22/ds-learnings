```python
# find the max frequency of the element in a tuple
tup = (1,3,2,1,4,4,4,4,4,2,5,5,3,7,6,3)
dic={}
for elem in tup:
    if elem not in dic:
        dic[elem] =1           #1:1, 3:1...
    else:
        dic[elem]+=1           #1:2,3:3...
max_freq  = 0
max_freq_elem=0
for key, val in dic.items():
    if val > max_freq:
        max_freq_elem = key
        max_freq = val
print(f"the max frequence elemnt is '{max_freq_elem}' occurs {max_freq} times")
```
